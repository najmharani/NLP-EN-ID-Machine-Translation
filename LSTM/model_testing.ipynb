```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

def encode_seq(tokenizer, maxlen, lines):
    seq = tokenizer.texts_to_sequences([lines])
    seq = pad_sequences(seq, maxlen=maxlen, padding='post')
    return seq
```

```python
import numpy as np

def decode_seq(tokenizer, prediction):
    seq = np.argmax(prediction[0], axis=-1)
    text = ' '.join([tokenizer.index_word.get(idx, '') for idx in seq if idx != 0])
    return text
```

```python
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction

def avg_bleu_score(input_data, output_data, input_tokenizer, output_tokenizer, max_sequence_length, model):
    bleu_scores = []
    results = []
    smoothing_function = SmoothingFunction().method1  # Apply smoothing

    for i in range(len(input_data)):
        
        # Tokenize and pad single sentence
        input_seq = encode_seq(input_tokenizer, 60, input_data[i])
        
        # Predict for this single input
        prediction = model.predict(input_seq)
        
        # Decode the prediction to text
        predicted_text = decode_seq(output_tokenizer, prediction)
        
        # Get the reference text
        reference_text = output_data[i]
        
        # Tokenize the reference and predicted sentences
        reference_tokens = [reference_text.split()]
        predicted_tokens = predicted_text.split()
        
        # Calculate BLEU score for this sentence
        bleu_score_value = sentence_bleu(reference_tokens, predicted_tokens, smoothing_function=smoothing_function)
        bleu_scores.append(bleu_score_value)

        # Save each input, reference, prediction, and BLEU score in the results list
        results.append({
            "input": input_data[i],
            "reference": reference_text,
            "prediction": predicted_text,
            "bleu_score": bleu_score_value
        })

    # Calculate the average BLEU score
    avg_bleu_score = np.mean(bleu_scores)
    print(f"Average BLEU Score: {avg_bleu_score:.4f}")

    return avg_bleu_score, results
    
```

```python
def save_model_comparisons(models, input_data, output_data, input_tokenizer, output_tokenizer, max_sequence_length, save_path):
    """
    Evaluate multiple models, save their results into separate files, and return average BLEU scores.
    
    Args:
        models (dict): A dictionary where keys are model names and values are model instances.
        input_data (list): List of input sentences.
        output_data (list): List of reference sentences.
        input_tokenizer: Tokenizer for the input language.
        output_tokenizer: Tokenizer for the output language.
        max_sequence_length (int): Maximum sequence length for padding.
        save_path (str): Path to save the combined results as an Excel file.
    
    Returns:
        dict: A dictionary where keys are model names and values are their average BLEU scores.
    """
    
    all_results = []
    avg_bleu_scores = {}
    comparison_dict = {"input": input_data, "reference": output_data}

    for model_name, model in models.items():
        print(f"Processing model: {model_name}")
        avg_bleu, model_results = avg_bleu_score(
            input_data, output_data, input_tokenizer, output_tokenizer, max_sequence_length, model
        )
        
        # Store average BLEU score
        avg_bleu_scores[model_name] = avg_bleu
        
        # Add model predictions to the comparison dictionary
        comparison_dict[f"{model_name}_prediction"] = [
            result["prediction"] for result in model_results
        ]
        
        # Add model name to each result for the results DataFrame
        for result in model_results:
            result["model"] = model_name
        
        all_results.extend(model_results)
    
    # Create the detailed results DataFrame
    results_df = pd.DataFrame(all_results)
    
    # Create the comparison DataFrame
    comparison_df = pd.DataFrame(comparison_dict)
    
    # Save both DataFrames to Excel
    results_save_path = save_path.replace(".xlsx", "_detailed_results.xlsx")
    comparison_save_path = save_path.replace(".xlsx", "_comparison.xlsx")
    
    results_df.to_excel(results_save_path, index=False)
    comparison_df.to_excel(comparison_save_path, index=False)
    
    print(f"Detailed results saved to {results_save_path}")
    print(f"Comparison table saved to {comparison_save_path}")
    
    return avg_bleu_scores, results_df, comparison_df
```

```python
def translate(line, input_tokenizer, output_tokenizer, maxlen, model):
    seq = encode_seq(input_tokenizer, maxlen, line)
    pred = model.predict(seq)
    trans = decode_seq(output_tokenizer, pred)
    
    return trans
```

```python
import pandas as pd

def compare_translations(models, input_data, input_tokenizer, output_tokenizer, max_sequence_length):
    """
    Compare translations from multiple models for the given input data.

    Args:
        models (dict): A dictionary where keys are model names and values are model instances.
        input_data (list): List of input sentences.
        input_tokenizer: Tokenizer for the input language.
        output_tokenizer: Tokenizer for the output language.
        max_sequence_length (int): Maximum sequence length for padding.

    Returns:
        pd.DataFrame: A DataFrame with inputs and predictions from all models.
    """
    comparison_dict = {"input": input_data}
    
    for model_name, model in models.items():
        print(f"Translating with model: {model_name}")
        comparison_dict[f"{model_name}_prediction"] = [
            translate(line, input_tokenizer, output_tokenizer, max_sequence_length, model)
            for line in input_data
        ]
    
    # Create a DataFrame from the comparison dictionary
    comparison_df = pd.DataFrame(comparison_dict)
    return comparison_df
```

```python
input_data = [
    "I love reading books.",
    "The weather is nice today.",
    "She is learning how to swim.",
    "We will visit the park tomorrow.",
    "Can you help me with my homework?",
    "Can we play now?"
]

comparison_df = compare_translations(
    models,
    input_data=input_data,
    input_tokenizer=en_tokenizer_100k,
    output_tokenizer=id_tokenizer_100k,
    max_sequence_length=60
)

# Save the comparison to a file
comparison_df.to_excel("model_predictions_comparison.xlsx", index=False)

# Display the DataFrame
print(comparison_df.head())
```
